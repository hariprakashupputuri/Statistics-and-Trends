```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
furniture_data = pd.read_csv('Furniture.csv')
```

```python
furniture_data
```

```python
# Generate descriptive statistics
furniture_describe = furniture_data.describe()
furniture_describe
```

```python
# Calculate Mean
mean_values = furniture_data.mean(numeric_only=True)
print("Mean:\n", mean_values)

# Calculate Median
median_values = furniture_data.median(numeric_only=True)
print("\nMedian:\n", median_values)

# Calculate Standard Deviation
std_dev_values = furniture_data.std(numeric_only=True)
print("\nStandard Deviation:\n", std_dev_values)

# Calculate Skewness
skewness_values = furniture_data.skew(numeric_only=True)
print("\nSkewness:\n", skewness_values)

# Calculate Kurtosis
kurtosis_values = furniture_data.kurtosis(numeric_only=True)
print("\nKurtosis:\n", kurtosis_values)
```

```python
# correlation matrix
furniture_corr = furniture_data.select_dtypes(include=['float64', 'int64']).corr()
print(furniture_corr)
```

```python
# Function to plot a histogram of a specified column
def plot_histogram(column_name):
    plt.figure(figsize=(8, 6))
    plt.hist(furniture_data[column_name], bins=20, edgecolor='black')
    plt.title(f"Distribution of {column_name.capitalize()}")
    plt.xlabel(column_name.capitalize())
    plt.ylabel("Frequency")
    plt.show()

# Plotting the histogram for the 'price' column
plot_histogram("price")
```

```python
# Function to plot scatter plot showing the relationship between price and revenue
def plot_scatter_price_vs_revenue():
    """
    Plots a scatter plot showing the relationship between price and revenue.
    Each point represents a furniture item, illustrating how price affects revenue.
    """
    plt.figure(figsize=(10, 6))
    plt.scatter(furniture_data['price'], furniture_data['revenue'], alpha=0.5, edgecolor='k')
    plt.title("Relationship between Price and Revenue")
    plt.xlabel("Price")
    plt.ylabel("Revenue")
    plt.grid(True)
    plt.show()

# Plotting the scatter plot for price vs revenue
plot_scatter_price_vs_revenue()
```

```python
 # Function to plot heatmap of the correlation matrix for numerical features
def plot_heatmap_correlation():
    """
    Plots a heatmap of the correlation matrix for numerical features in the furniture dataset.
    This visual representation helps identify strong positive or negative relationships.
    """
    # Calculate correlation matrix within the function
    furniture_corr = furniture_data.corr(numeric_only=True)
    
    plt.figure(figsize=(8,6))
    sns.heatmap(furniture_corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
    plt.title("Correlation Matrix Heatmap")
    plt.show()

# Plotting the heatmap for the correlation matrix
plot_heatmap_correlation()
```
